```python
import pandas as pd
dialect_df = pd.read_csv('/content/drive/MyDrive/arabic_dialects_clean.csv')
dialect_df.head()
```

```python
maghrebi_df = dialect_df[dialect_df['dialect'].isin(['TN','DZ','MA','LY'])]
```

```python
# filter the dataframe to keep only the desired dialects
maghrebi_df = dialect_df[dialect_df['dialect'].isin(['TN', 'DZ', 'MA', 'LY'])]

# reduce the number of rows for the 'MA' dialect to 37,000
maghrebi_df.loc[maghrebi_df['dialect'] == 'MA', 'dialect'] = maghrebi_df[maghrebi_df['dialect'] == 'MA'].sample(n=40000, random_state=42)['dialect']

# reduce the number of rows for the 'TN' dialect to 37,000
maghrebi_df.loc[maghrebi_df['dialect'] == 'TN', 'dialect'] = maghrebi_df[maghrebi_df['dialect'] == 'TN'].sample(n=40000, random_state=42)['dialect']
maghrebi_df['dialect'].value_counts()
```
